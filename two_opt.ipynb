```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
cities = 30
pos = np.random.rand(cities,2)
distance = {}
```

```python
for i,posi in enumerate(pos):
    for j,posj in enumerate(pos):
        if j>i:
            key = (i,j)
            distance[key] = math.sqrt(sum((posi-posj)**2))
            #print('{} : {}'.format(key,distance[key]))
```

```python
def dot_swap(pre_pos,index1,index2):
    new_pos = pre_pos.copy()
    if isinstance(pre_pos , np.ndarray):
        tmp = new_pos[index1].copy()
        new_pos[index1] = new_pos[index2].copy()
        new_pos[index2] = tmp
    return new_pos
```

```python
def calcDistance(pos):
    totalDistance = 0
    for i in range(len(pos)):
        if i < len(pos)-1:
            totalDistance += math.sqrt(sum( (pos[i]-pos[i+1]) ** 2 ))
    return totalDistance + math.sqrt(sum( (pos[i]-pos[0]) ** 2 ))
```

```python
def two_opt(pos): 
    best_pos = pos
    tag = True
    while tag:
        tag = False
        for i in range(0, len(pos)):
            for j in range(i, len(pos)):
                new_pos = dot_swap(best_pos,i,j)
                if calcDistance(new_pos) < calcDistance(pos):
                    #print(calcDistance(pos))
                    best_pos = new_pos.copy()
                    tag = True
        pos = best_pos
    return best_pos
```

```python
plt.figure(0)
plt.title('Graph TSP')
plt.scatter(pos[:,0], pos[:,1])
plt.plot(pos[:,0],pos[:,1],'b--')
bestEver = two_opt(pos)
plt.figure(1)
plt.title('Graph TSP - TwoOpt')
plt.scatter(pos[:,0], pos[:,1])
plt.plot(bestEver[:,0],bestEver[:,1],'r--')
```
